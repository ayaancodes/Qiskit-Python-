# Introduction to Quantum Circuits
**Objective:** Learn how to create quantum circuits, apply quantum gates, and simulate them.

```python
import numpy as np

# Qiskit imports:
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
import qiskit.quantum_info as qi
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

```

## Step 1: Creating a Qubit Circuit

```python
# Create a new circuit with one qubit
qc = QuantumCircuit(1)

# Draw the circuit
qc.draw("mpl")
```

## Step 2: Adding Measurement

```python
# Add a measurement device to measure the qubit
qc.measure_all()

# Draw the updated circuit
qc.draw("mpl")
```

## Step 3: Simulating the Circuit
Now that we have created the circuit, we will **simulate** it using Qiskit's AerSimulator.

```python
# Set up the Aer simulator
simulator = AerSimulator()

# Transpile the circuit for the simulator
qc = transpile(qc, simulator)

# Run the circuit for 10 shots
result = simulator.run(qc, shots=10, memory=True).result()

# Get the measurement results
memory = result.get_memory(qc)

# Print the output
print(memory)
```

## Step 4: Applying the Pauli-X Gate
- The X gate flips the qubit from |0⟩ to |1⟩.
- After applying X and measuring, all outputs should be '1'.

```python
# Create a circuit with one qubit
qc = QuantumCircuit(1)

# Apply X gate
qc.x(0)

# Add measurement
qc.measure_all()

# Draw the circuit
qc.draw("mpl")
```

```python
# Simulate
simulator = AerSimulator()
qc = transpile(qc, simulator)

result = simulator.run(qc, shots=10, memory=True).result()
memory = result.get_memory(qc)
print(memory)
```

## Step 5: Hadamard Gate (Superposition)
- The H gate puts the qubit in superposition, giving 50% probability of |0⟩ and |1⟩.
- After measurement, results should be a mix of '0' and '1'.


```python
# Create a new circuit with one qubit
qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()
# Transpile for simulator
simulator = AerSimulator()
qc = transpile(qc, simulator)
# Run the circuit for 10 trials, or "shots", and print out the results
result = simulator.run(qc,shots=10,memory=True).result()
memory = result.get_memory(qc)
print(memory)
```

```python
# Run and get counts in a histogram
result = simulator.run(qc,shots=10).result()
counts = result.get_counts(qc)
plot_histogram(counts, title='State counts')
```

```python
# Run and get counts in a histogram
result = simulator.run(qc,shots=100).result()
counts = result.get_counts(qc)
plot_histogram(counts, title='State counts')
```

# Question 1: The Hadamard Gate and Measurement Statistics

## Q1.1: Experiment with Different Numbers of Shots
- We measure the qubit after applying a Hadamard gate.
- The experiment is repeated with different numbers of shots: **10, 100, 1000, 10,000, and 100,000**.
- We extract the probability of measuring `|0⟩` and compare it to the expected 50%.


```python
# Create a simple quantum circuit (e.g., Hadamard on a single qubit)
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)

# Initialize simulator
simulator = AerSimulator()

# Define shot values
shot_values = [10, 100, 1000, 10_000, 100_000]
errors = []

# Theoretical probabilities for a Hadamard gate (equal superposition)
p_0_theory = 0.5
p_1_theory = 0.5

for shots in shot_values:
    # Run the circuit
    result = simulator.run(qc, shots=shots).result()
    counts = result.get_counts(qc)

    # Get measured probabilities
    p_0_measured = counts.get('0', 0) / shots
    p_1_measured = counts.get('1', 0) / shots

    # Calculate absolute errors
    error_0 = abs(p_0_measured - p_0_theory)
    error_1 = abs(p_1_measured - p_1_theory)

    # Store average error
    errors.append((error_0 + error_1) / 2)

# Convert errors and shots to NumPy arrays for plotting
shot_values = np.array(shot_values)
errors = np.array(errors)

# Theoretical error scaling (1/sqrt(N))
expected_errors = 1 / np.sqrt(shot_values)

# Plot results
plt.figure(figsize=(12, 5))

# Regular plot
plt.subplot(1, 2, 1)
plt.plot(shot_values, errors, 'o-', label="Measured Error")
plt.plot(shot_values, expected_errors, 'r--', label="1/√N Trend")
plt.xlabel("Number of Shots (N)")
plt.ylabel("Absolute Error")
plt.title("Error vs. Number of Shots (Linear Scale)")
plt.legend()

# Log-log plot
plt.subplot(1, 2, 2)
plt.loglog(shot_values, errors, 'o-', label="Measured Error")
plt.loglog(shot_values, expected_errors, 'r--', label="1/√N Trend")
plt.xlabel("Number of Shots (log N)")
plt.ylabel("Absolute Error (log scale)")
plt.title("Error vs. Number of Shots (Log-Log Scale)")
plt.legend()

plt.show()
```

### 

### **Q1.2: Plotting the Error vs. N**

The absolute error is analyzed as a function of the number of shots, $ N $, to observe how measurement accuracy improves with increased sampling. 

Two different plots are generated:
- **Linear Scale Plot**: Shows how the absolute error decreases as $ N $ increases.
- **Log-Log Scale Plot**: Highlights the expected **inverse square root** relationship,  $ \frac{1}{\sqrt{N}} $, between error and $ N $.

---

#### **1. Linear Scale Analysis**
- The absolute error shows a **sharp decrease** for smallerf $ N $, particularly between \( N = 10 \) and \( N = 100 \).
- As \( N \) increases further, the error reduction slows down and approaches a stable value.
- This behavior follows the expected trend where larger samples yield more accurate probability estimates.

#### **2. Log-Log Scale Analysis**
- On a **log-log scale**, a relationship following  $ \frac{1}{\sqrt{N}} $ should appear as a **straight line** with a slope of approximately **-0.5**.
- The measured data aligns well with this theoretical expectation.
- Small deviations are visible at intermediate values of $ N $, which can be attributed to statistical noise.

---

### **Explanation of the Trend  $ \frac{1}{\sqrt{N}} $**
1. **Statistical Fluctuations**  
   - At small \( N \), random measurement fluctuations can cause noticeable deviations from the expected trend.
   - These variations decrease as $ N $ grows, reducing the sampling error.

2. **Law of Large Numbers**  
   - As the number of shots increases, the observed probability of measuring $ |0\rangle $ converges to the true probability of **50%**.
   - This results in a decreasing error as predicted by  $ \frac{1}{\sqrt{N}} $.



# Question 2: Generating Arbrtrary Single Qubit State


```python
qc = QuantumCircuit(1)

# Apply RY(π/3) gate
qc.ry(np.pi/3, 0)

# Apply P(π/2) gate
qc.p(np.pi/2, 0)

# Measure the qubit
qc.measure_all()

# Draw the circuit
qc.draw("mpl")
```

### **Q2.1: Theory**

By applying the sequence  
$$ P(\gamma)\, R_Y(\theta)\,|0\rangle $$ we can construct **any valid** single-qubit state (up to an overall global phase).  

1.  The gate $ R_Y(\theta) $  rotates the qubit's state vector around the Y-axis by an angle  $ \theta $.  
2. Then $ P(\gamma) $ (the phase gate) adds a relative phase $ e^{i\gamma} $ between $ |0\rangle $ and $ |1\rangle $.  


The corresponding matrices are:

$$
R_Y(\theta)
=
\begin{bmatrix}
\cos(\tfrac{\theta}{2}) & -\sin(\tfrac{\theta}{2}) \\
\sin(\tfrac{\theta}{2}) & \cos(\tfrac{\theta}{2})
\end{bmatrix}
,\quad
P(\gamma)
=
\begin{bmatrix}
1 & 0 \\
0 & e^{i\gamma}
\end{bmatrix}.
$$

Applying these transformations sequentially to $ |0\rangle $ results in the following general qubit state:

$$
\cos\Bigl(\tfrac{\theta}{2}\Bigr)|0\rangle \;+\;
e^{i\gamma}\,\sin\Bigl(\tfrac{\theta}{2}\Bigr)|1\rangle.
$$

This demonstrates that any quantum state on the Bloch sphere can be reached using these two gates, apart from an unobservable global phase.

### **Why Order Matters**
- The phase gate $ P(\gamma) $ and the rotation gate  $ R_Y(\theta) $ **do not commute**.
- $ P(\gamma) $ only applies a phase shift to the $ |1\rangle $ state, but if it is applied **before** $ R_Y(\theta) $, the qubit remains in $ |1\rangle $, making $ P(\gamma) $ ineffective.
- Instead, applying $ R_Y(\theta)$ first creates a superposition, allowing $ P(\gamma) $ to meaningfully introduce the phase shift.

## **Question 3: Comparing Theoretical and Experimental Probabilities**

### **Q3.1: Expected Probabilities from Theory**

Applying the sequence  
$$ P\bigl(\tfrac{\pi}{2}\bigr)\, R_Y\bigl(\tfrac{\pi}{3}\bigr) |0\rangle $$  
transforms the qubit state into:  
$$
\cos\bigl(\tfrac{\pi}{6}\bigr)|0\rangle 
\;+\;
e^{i \frac{\pi}{2}}
\sin\bigl(\tfrac{\pi}{6}\bigr)|1\rangle.
$$  

Since  
$$
\cos\Bigl(\tfrac{\pi}{6}\Bigr) = \frac{\sqrt{3}}{2},
\quad
\sin\Bigl(\tfrac{\pi}{6}\Bigr) = \frac{1}{2},
$$  
we express the state as:  
$$
\frac{\sqrt{3}}{2} |0\rangle + i\frac{1}{2} |1\rangle.
$$  

From this, we determine the theoretical measurement probabilities:

So the probabilities are:  
$$
P(0) = |a|^2 = \left(\frac{\sqrt{3}}{2}\right)^2 = \frac{3}{4} = 0.75,
\quad
P(1) = |b|^2 = \left(\frac{1}{2}\right)^2 = \frac{1}{4} = 0.25.
$$

Thus, in an ideal noiseless simulation, the qubit should be measured in the \( |0\rangle \) state **75% of the time** and in the \( |1\rangle \) state **25% of the time**.
